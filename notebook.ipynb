```r
sales_summary <- sales_clean %>%
group_by (sales_method) %>%
summarize(median= median (revenue, na.rm=TRUE), average=mean(revenue, na.rm=TRUE))

sales_summary
```

Before I decide if I want to use the median or average for imputation I want to plot the distribution of revenue by sales_method using ggplot2.

Because all of these are skewed we will use the medians for imputation.

```r
#Impute medians by sales_method
sales_clean <- sales_clean %>%
group_by(sales_method) %>%
mutate(revenue= if_else(is.na(revenue), median(revenue, na.rm = TRUE), revenue)) %>%
ungroup()

#confirm no null values
sales_clean %>%
filter(is.na(revenue))
```

The fact that we have no results means that there are no NAs left and the column has been successfully cleaned.

Now I can do some hypothesis testing to see if there's any significant difference in sales_method and revenue or nb_sold. Let's start with revenue. Because this involves categorical data and 3 different treatment groups we cannot use the t.test or chi squared test. We will therefore use analysis of variance

```r
AOV <- aov(sales_clean$revenue ~ sales_clean$sales_method, sales_clean)

summary(AOV)
```

There is a significant difference in revenue between sales methods. To figure out which ones are significant we need run a t test on the two methods with the highest average revenue.

```r
# Reference sales_summary to determine which groups to test
sales_summary

# run a t.test on "Email" and "Email + Call"
## first filter the data to just those two groups
ttest_data<- sales_clean %>%
filter(sales_method == "Email" | sales_method=="Email + Call")

#confirm call has been removed
ttest_data %>%
count(sales_method)



##Preform t.test
t.test(ttest_data$revenue ~ ttest_data$sales_method, data= ttest_data, lower.tail=FALSE)
```

The p-value for this is below 0.05 so it is a significant difference. 

The best sales method in terms of revenue is to do an email and a call.
